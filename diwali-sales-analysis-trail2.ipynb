```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/kaggle/input/diwali-sales-analysistrial2/Diwali Sales Data.csv", encoding="unicode_escape")
df
# Clean the data
df.dropna(subset=['Marital_Status', 'Age Group', 'Gender', 'State', 'Occupation', 'Product_Category', 'Amount'], inplace=True)  # Ensure no missing values
df['Marital_Status'] = df['Marital_Status'].astype(str)  # Convert to string if necessary


df.head()
# Get the shape of the DataFrame (rows, columns)
df.shape  

# Count of missing values in each column
df.isnull().sum()  

# Summary statistics for numerical columns
df.describe()  


# Check if DataFrame is empty before plotting
if df.empty:
    print("Error: The dataset is empty after cleaning!")
else:
    # Countplot for marital status
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Marital_Status', data=df, palette='Set2')
    plt.title("Marital Status Distribution")
    plt.xlabel("Marital Status")
    plt.ylabel("Count")
    plt.show()

    # Age Group-wise Sales
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Age Group', data=df, hue='Gender', palette='Set2')
    plt.title("Age Group-wise Sales")
    plt.xlabel("Age Group")
    plt.ylabel("Count")
    plt.legend(title="Gender")
    plt.show()

    # Total Amount by Age Group
    age_amount = df.groupby('Age Group')['Amount'].sum().reset_index()
    plt.figure(figsize=(6, 4))
    sns.barplot(x='Age Group', y='Amount', data=age_amount, palette='Blues')
    plt.title("Total Amount Spent by Age Group")
    plt.xlabel("Age Group")
    plt.ylabel("Total Amount Spent")
    plt.show()

    # Marital Status and Sales
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Marital_Status', hue='Gender', data=df, palette='Set3')
    plt.title("Marital Status-wise Sales")
    plt.xlabel("Marital Status")
    plt.ylabel("Count")
    plt.legend(title="Gender")
    plt.show()

    # State-wise Sales
    state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False).reset_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Amount', y='State', data=state_sales, palette='coolwarm')
    plt.title("State-wise Total Sales")
    plt.xlabel("Total Sales Amount")
    plt.ylabel("State")
    plt.show()

    # Occupation-wise Sales
    occ_sales = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Amount', y='Occupation', data=occ_sales, palette='viridis')
    plt.title("Occupation-wise Sales")
    plt.xlabel("Total Sales Amount")
    plt.ylabel("Occupation")
    plt.show()

    # Product Category Sales
    prod_sales = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index()
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Amount', y='Product_Category', data=prod_sales, palette='muted')
    plt.title("Sales by Product Category")
    plt.xlabel("Total Sales Amount")
    plt.ylabel("Product Category")
    plt.show()

    # Boxplot to Analyze Spending Patterns
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Age Group', y='Amount', data=df, palette='pastel')
    plt.title("Spending Patterns by Age Group")
    plt.xlabel("Age Group")
    plt.ylabel("Amount Spent")
    plt.show()

    # Correlation Heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title("Correlation Heatmap")
    plt.show()
```

**#Diwali Sales Data Analysis Report**

**1. Which gender shops the most?**
* **Analysis:**
From the countplot, we see that females have made more purchases than males.
However, from the total amount spent (bar plot), males tend to spend more on average per transaction than females.
* **Conclusion:**
* *Women make more purchases, but men spend more per transaction.*

**2. Which age group shops the most?**
* **Analysis:**

  The 26-35 years age group has the highest number of purchases.The total spending (bar plot) is also highest for this age group.

* **Conclusion:**
  Young adults aged 26-35 are the biggest shoppers in terms of both the number of purchases and total amount spent.

**3. Which state generates the highest revenue?**
* **Analysis:**
  From the state-wise total sales bar chart, we observe that Maharashtra, Uttar Pradesh, and Karnataka contribute the highest sales revenue.
* **Conclusion:**
 Maharashtra has the highest total sales, followed by Uttar Pradesh and Karnataka.

**4. Which occupation spends the most?**
* **Analysis:**
The bar plot of occupation-wise sales shows that IT professionals and Healthcare workers spend the most.
* **Conclusion:**
IT professionals are the highest spenders, likely due to higher disposable income.

**5. Which product categories generate the highest revenue?**
* **Analysis:**
The top-selling product categories in terms of revenue are Electronics, Clothing, and Footwear.
* **Conclusion:**
Electronics generate the most revenue, followed by Clothing and Footwear

**6. What is the spending pattern across different age groups?**
* **Analysis:**
A box plot of Age Group vs. Amount Spent shows that older customers (46-55 years) tend to make larger purchases occasionally.
The younger group (18-25) has lower spending per purchase.
